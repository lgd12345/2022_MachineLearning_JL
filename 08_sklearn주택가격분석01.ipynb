```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston = load_boston()
print(boston)
```

```python
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()
```

```python
boston_df['PRICE']= boston.target
boston_df.head()
```

```python
print("보슨톤 주택 가격 데이터 셋 크기: ",boston_df.shape)
#타켓하나 더 추가해서 14가 됨
```

```python
boston_df.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score #에러의 제곱의 평균
```

```python
#X, Y 분할하기
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'], axis = 1, inplace = False)
```

```python
#훈련용 데이터와 평가용 데이터 분할하기 70:30
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, 
random_state = 156)
```

```python
#선형 회귀 분석 : 모델 생성
lr = LinearRegression()
```

```python
#선형 회귀 분석 : 모델 훈련
lr.fit(X_train, Y_train)
```

```python
#선형 회귀 분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
print(Y_predict)
```

```python
print(Y_test)
```

```python
mse = mean_squared_error(Y_test, Y_predict) #에러 제곱 평균
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index = X.columns)#시리즈를 만들고
coef.sort_values(ascending = False)#역순으로 보여주기
```
