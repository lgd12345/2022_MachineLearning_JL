```python
import pandas as pd
```

```python
file_name='코로나_naver_news'
data_df=pd.read_csv('data/코로나_naver_news.csv',encoding='utf-8')
```

```python
data_df.head()
```

```python
# 시리즈 인덱스 + 칼럼
description = data_df['description']
description
```

```python
from konlpy.tag import Okt
okt=Okt()
```

```python
description_noun_tk = []
for d in description:
    description_noun_tk.append(okt.nouns(d)) #명사 형태소만 추출
```

```python
#토큰들 확인
description_noun_tk
```

```python
description_noun_tk2 = []
for d in description_noun_tk:
    item = [i for i in d if len(i) > 1] #토큰 길이가 1보다 큰 것만 추출
    description_noun_tk2.append(item)
        
#  위에랑 같은 문장이다.        
#     item = []
#     for i in d:
#         if len(i)>1:
#             item.append(i)
```

```python
print(description_noun_tk2)
```

```python
#(lda모델 구축)
import gensim
import gensim.corpora as corpora
```

```python
# 토픽을 위한 단어사전 생성 
dictionary = corpora.Dictionary(description_noun_tk2)
print(dictionary[0]) #작업 확인용 출력
print(dictionary[1])
print(dictionary[100])
```

```python
# 단어와 출현 빈도에 코퍼스 생성 17번 단어 2번 .. 이런식
corpus = [dictionary.doc2bow(word) for word in description_noun_tk2]
```

```python
print(corpus)
```

```python
k = 4 #토픽의 개수 설정
```

```python
lda_model = gensim.models.ldamulticore.LdaMulticore(corpus, iterations = 12, 
                                                    num_topics = k,
                                                    id2word = dictionary,
                                                    passes = 1,
                                                    workers = 10)
#curpus: 단어(id)별 출현 빈도
#id2word: id to word
#num_topice: 생성될 토픽의 개수
#workers: 한번의 트레이닝에 처리될 문서의 개수
#passes: 전체 코퍼스 트레이닝 횟수
#interations: 문서 당 반복 횟수
```

```python
# 15에 따라 토픽을 구성하는 주요 단어 15개가 토픽에 대한 영향력 비율 과 함께 출력된 것을 확인
# 네이버 뉴스를 크롤링할 때 검색어로 ‘코로나’를 사용했기 때문에 모든 토픽에서 ‘코로나’ 단어가 압도적으로 많이 나옴

# 토픽 분석 결과 출력
print(lda_model.print_topics(num_topics = k, num_words = 15))
```
